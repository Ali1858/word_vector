```python
import numpy as np
import codecs
import pandas as pd
import csv
import re

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
import gensim.models as gm
```

```python
file = 'news_source.txt'
#glove vector with 50 dimension
glove_vect_50d = 'glove.6B/glove.6B.50d.txt'
#glove vector with 300 dimension
glove_vect_300d = 'glove.6B/glove.6B.300d.txt'
#word2vec with 300 dimension
word2vec_path = 'GoogleNews-vectors-negative300.bin/GoogleNews-vectors-negative300.bin'
corpus = u""
with codecs.open(file,"r","utf-8") as news_file:
    corpus = news_file.read()
```

```python
#split the sentences 
raw_sentences  = corpus.split(".")
```

```python
# remove all un wanted bracket special caractor etc
sentences = []
for raw_sentence in raw_sentences:
    if len(raw_sentence)>0:
        clean = re.sub("[^a-zA-Z0-9’]"," ",raw_sentence)
        sentences.append(clean.split())
        
        
```

```python
#split each sentences into words
vocab = []
for each_sentences in sentences:
    for words in each_sentences:
        vocab.append(words)
#and select unique words
vocab = set(vocab)
```

```python
#glove vector of 50 and 300 dimensions from txt file
#link for download https://nlp.stanford.edu/projects/glove/
glove_50D= pd.read_table(glove_vect_50d, sep=" ", index_col=0, header=None, quoting=csv.QUOTE_NONE)
glove_300D = pd.read_table(glove_vect_300d, sep=" ", index_col=0, header=None, quoting=csv.QUOTE_NONE)
```

```python
glove_50D.head()
```

```python
glove_300D.head()
```

```python
word_vec_glove50D = []
word_vec_glove300D = []
found_vocab = []
unfound_word_glove= []
for voc in vocab:
    #check if vocab is available in vector
    #if yes value of latent features and append in our vector list
    if voc in glove_50D.index:
        word_vec_glove50D.append(glove_50D.loc[voc].as_matrix())
        word_vec_glove300D.append(glove_300D.loc[voc].as_matrix())
        found_vocab.append(voc)
    else:
        unfound_word_glove.append(voc)
```

```python
print('dimensions for glove vector with 50 latent featuresn')
print(glove_50D.shape,len(word_vec_glove50D))
print('\ndimensions for glove vector with 300 latent featuresn')
print(glove_300D.shape,len(word_vec_glove300D))
print('\nunfound words in glove vector')
print(len(unfound_word_glove))
```

```python

def plot_graph(data,vocab):
    if len(data) == 0:
        print("No data in this region")
        return
    plt.scatter(data[:, 0], data[:, 1],s=10)
    for label, x, y in zip(vocab, data[:, 0], data[:, 1]):
        plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
    plt.show()
    print('done')
```

```python
def plot_region(x_bound,y_bound,datas,vocab):
    slice_word = []
    slice_data = []
    for word,data in zip(vocab,datas):
        
        if x_bound[0]<= data[0] and data[0]<= x_bound[1] and y_bound[0]<=data[1] and data[1]<=y_bound[1]:
            slice_word.append(word)
            slice_data.append(data)
    
    plot_graph(np.array(slice_data),slice_word)

    
```

```python
#dimenstionality reduction using TSNE for 50D and 300D 
tsne =TSNE(n_components=2,random_state=0)
glove_2D_50 = tsne.fit_transform(word_vec_glove50D)

glove_2D_300 = tsne.fit_transform(word_vec_glove300D)
```

```python
plot_graph(glove_2D_300,found_vocab)
```

```python
#slice of region 
plot_region([5,10],[2,8],glove_2D_300,found_vocab)
```

```python
plot_graph(glove_2D_50,found_vocab)
```

```python
plot_region([5,20],[-15,1],glove_2D_50,found_vocab)
```
